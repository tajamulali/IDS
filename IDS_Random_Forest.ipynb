```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score
```

```python
df = pd.read_csv('C:/Users/hp/OneDrive/Desktop/MCA/Second_Sem/Data_Sets/train_data_1.csv')
df.head()
```

```python
print(df['protocol_type'].unique())
df['xAttack'].unique()
```

```python
proto_dict = {
    'tcp':0,
    'udp':1,
    'icmp':2
}
proto_dict_1 = {
    'normal':0,
    'dos':1,
    'r2l':1,
    'probe':1,
    'u2r':1
}

df['protocol_type']=df['protocol_type'].map(proto_dict)
df['xAttack']=df['xAttack'].map(proto_dict_1)
```

```python
df.head()
```

```python
empty_row = df.isnull().all(axis=0)
empty_col= df.isnull().all(axis=1)
print(empty_row)
print(empty_col)
```

```python
y = df['xAttack']
print(y)
```

```python
X = df.drop('xAttack', axis=1)
print(X)
```

```python
Train_X, Val_x, Train_y, Val_y = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
df_model = RandomForestClassifier()
df_model.fit(Train_X,Train_y)

prediciton = df_model.predict(Val_x)
```

```python
print(prediciton)
```

```python
print(classification_report(Val_y,prediciton))
```

```python
print(accuracy_score(Val_y,prediciton))
```
